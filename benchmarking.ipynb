## Imports

```python
import matplotlib.pyplot as plt
import graphs
import time
```

## Loading Data

```python
train_graph, test_graph =graphs.load_data("facebook_data\\training_set.tsv", "facebook_data\\testing_set.tsv")
```

## Benchmarking

```python
num_edges = []
for i in range(10):
    G = graphs.recommend_all_friends(train_graph, i + 1)
    num_edges.append(len(G._edges))
```

```python
num_edges = []
times = []
for i in range(10):
    start_time = time.perf_counter()
    G = graphs.recommend_all_friends(train_graph, i + 1)
    end_time = time.perf_counter()
    print(f"At depth {i + 1} the number of edges: {G.count_edges()}")
    num_edges.append(G.count_edges())
    times.append(end_time - start_time)
```

## Graphing

```python
plt.scatter(range(1, 11), num_edges)
plt.xlabel("Max Depth")
plt.ylabel("# of edges in Graph")
```

```python
plt.scatter(range(1, 11), times)
plt.xlabel("Max Depth")
plt.ylabel("Runtime (seconds)")
```

## Accuracy Evaluation

```python
precision = []
recall = []
for i in range(10):
    G = graphs.recommend_all_friends(train_graph, i + 1)
    precision.append(graphs.precision(G, test_graph))
    recall.append(graphs.recall(G, test_graph))
```

## Graphing Evaluation

```python
plt.scatter(range(1, 11), precision)
plt.xlabel("Max Depth")
plt.ylabel("Precision")
```

```python
plt.scatter(range(1, 11), recall)
plt.xlabel("Max Depth")
plt.ylabel("Recall")
```
